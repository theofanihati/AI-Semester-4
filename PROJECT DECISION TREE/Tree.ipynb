```python
import pandas as pd #manipulasi dan analisis data
df = pd.read_csv("cuaca.csv")
df
```

```python
inputs = df.drop('bermain', axis='columns')
target = df['bermain']
```

```python
target
```

```python
#mengubah label (kategori) menjadi angka
from sklearn.preprocessing import LabelEncoder 
```

```python
le_Pandangan = LabelEncoder()
le_Suhu = LabelEncoder()
le_Kelembapan = LabelEncoder()
```

```python
inputs ['Pandangan_n'] = le_Pandangan.fit_transform(inputs['Pandangan'])
inputs ['Suhu_n'] = le_Suhu.fit_transform(inputs['Suhu'])
inputs ['Kelembapan_n'] = le_Kelembapan.fit_transform(inputs['Kelembapan'])
inputs
```

```python
inputs_n = inputs.drop(['Pandangan', 'Suhu', 'Kelembapan'], axis='columns')
inputs_n
```

```python
from sklearn import tree
```

```python
model = tree.DecisionTreeClassifier()
```

```python
model.fit(inputs_n,target)
```

```python
model.score(inputs_n,target)
```

```python
model.predict([[2,2,1]])
```
